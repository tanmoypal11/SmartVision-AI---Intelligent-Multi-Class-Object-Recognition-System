```python
# ============================================================
# SMARTVISION AI - FINAL STABLE TRAINING CODE
# ============================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import os

# 1. Device Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 2. DEFINITIVE PATHS (Based on your os.listdir results)
DATASET_DIR = "/content/dataset/classification"

# Model saving path on Drive
BASE_DRIVE_DIR = "/content/drive/MyDrive/SmartVision AI - Intelligent Multi-Class Object Recognition System"
models_dir = f"{BASE_DRIVE_DIR}/models"
os.makedirs(models_dir, exist_ok=True)
best_model_path = f"{models_dir}/best_mobilenetv2_smartvision.pth"

# 3. Data Loaders (SSD Optimized)
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load Datasets
train_dataset = ImageFolder(f"{DATASET_DIR}/train", transform=train_transform)
val_dataset   = ImageFolder(f"{DATASET_DIR}/val",   transform=val_test_transform)
test_dataset  = ImageFolder(f"{DATASET_DIR}/test",  transform=val_test_transform)

# DataLoader with 4 workers for speed
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=32, shuffle=False, num_workers=4, pin_memory=True)
test_loader  = DataLoader(test_dataset,  batch_size=32, shuffle=False, num_workers=4, pin_memory=True)

print(f"âœ… Success! Found {len(train_dataset.classes)} classes: {train_dataset.classes}")

# 4. Build Model (MobileNetV2)
model = models.mobilenet_v2(weights="IMAGENET1K_V1")

# Freeze early layers
for param in model.parameters():
    param.requires_grad = False

# Replace classifier head
model.classifier[1] = nn.Linear(model.classifier[1].in_features, len(train_dataset.classes))

# Unfreeze last 2 blocks for fine-tuning
for block in model.features[16:]:
    for param in block.parameters():
        param.requires_grad = True

model = model.to(device)

# 5. Optimization Utilities
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam([
    {"params": model.features[16:].parameters(), "lr": 1e-4}, # slow learning for base
    {"params": model.classifier.parameters(),    "lr": 1e-3} # faster learning for new head
])
scaler = torch.amp.GradScaler('cuda')
history = {'train_acc': [], 'val_acc': []}

# 6. Training Loop
num_epochs = 15
best_acc = 0.0

print("\n" + "="*40)
print("ðŸš€ TRAINING STARTED ON LOCAL SSD")
print("="*40)

try:
    for epoch in range(num_epochs):
        model.train()
        train_correct, train_total = 0, 0

        loop = tqdm(train_loader, desc=f"Epoch [{epoch+1}/{num_epochs}]", leave=True)
        for images, labels in loop:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            with torch.amp.autocast('cuda'):
                outputs = model(images)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            _, preds = outputs.max(1)
            train_total += labels.size(0)
            train_correct += preds.eq(labels).sum().item()
            loop.set_postfix(loss=f"{loss.item():.4f}")

        # Validation Phase
        model.eval()
        val_correct, val_total = 0, 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, preds = outputs.max(1)
                val_total += labels.size(0)
                val_correct += preds.eq(labels).sum().item()

        val_acc = 100 * val_correct / val_total
        history['train_acc'].append(100 * train_correct / train_total)
        history['val_acc'].append(val_acc)

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), best_model_path)
            status = "ðŸŒŸ BEST!"
        else:
            status = ""

        print(f"Summary Epoch {epoch+1}: Train: {history['train_acc'][-1]:.2f}% | Val: {val_acc:.2f}% {status}")

except KeyboardInterrupt:
    print("\nTraining paused by user.")

# 7. Final Results Visualization
plt.figure(figsize=(10, 5))
plt.plot(history['train_acc'], label='Train Accuracy', marker='o')
plt.plot(history['val_acc'], label='Val Accuracy', marker='o')
plt.title('SmartVision AI Training Progress')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from tqdm.auto import tqdm
import os
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score

# -----------------------------
# Device & Paths
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
base_path = "/content/dataset/classification"
train_dir, val_dir, test_dir = [os.path.join(base_path, d) for d in ["train", "val", "test"]]

models_dir = "/content/drive/MyDrive/SmartVision AI - Intelligent Multi-Class Object Recognition System/models"
os.makedirs(models_dir, exist_ok=True)
best_model_path = os.path.join(models_dir, "best_efficientnet_high_capacity.pth")

# -----------------------------
# 1. Transforms
# -----------------------------
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(0.2, 0.2, 0.2, 0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# -----------------------------
# 2. Datasets & Loaders
# -----------------------------
train_dataset = ImageFolder(train_dir, transform=train_transform)
val_dataset   = ImageFolder(val_dir, transform=val_test_transform)
test_dataset  = ImageFolder(test_dir, transform=val_test_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True)

# -----------------------------
# 3. Model Setup (Fine-Tuning Block 4+)
# -----------------------------
model = models.efficientnet_b0(weights="IMAGENET1K_V1")

for param in model.parameters():
    param.requires_grad = False

# Unfreeze from Block 4 onwards to accommodate larger data (2500/class)
for block in model.features[4:]:
    for param in block.parameters():
        param.requires_grad = True

model.classifier[1] = nn.Linear(model.classifier[1].in_features, len(train_dataset.classes))
model = model.to(device)

# -----------------------------
# 4. Optimization
# -----------------------------
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.AdamW([
    {"params": model.features[4:].parameters(), "lr": 1e-4},
    {"params": model.classifier.parameters(), "lr": 1e-3}
], weight_decay=0.01)

num_epochs = 20
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

# -----------------------------
# 5. Training Loop
# -----------------------------
best_val_acc = 0.0
print(f"\nðŸš€ Training with {len(train_dataset)} images...")

for epoch in range(num_epochs):
    model.train()
    t_correct, t_total = 0, 0
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", leave=False)

    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        _, preds = outputs.max(1)
        t_total += labels.size(0)
        t_correct += preds.eq(labels).sum().item()

    model.eval()
    v_correct, v_total = 0, 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = outputs.max(1)
            v_total += labels.size(0)
            v_correct += preds.eq(labels).sum().item()

    val_acc = 100 * v_correct / v_total
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), best_model_path)
        print(f"\nðŸŒŸ New Best: {val_acc:.2f}% (Saved)")

    print(f"Epoch [{epoch+1}/{num_epochs}] | Train Acc: {100*t_correct/t_total:.2f}% | Val Acc: {val_acc:.2f}%")
    scheduler.step()

# -----------------------------
# 6. Advanced Evaluation & Metrics
# -----------------------------
print("\n" + "="*50)
print("GENERATING DETAILED PERFORMANCE METRICS")
print("="*50)

model.load_state_dict(torch.load(best_model_path))
model.eval()

all_preds, all_labels, inference_times = [], [], []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        start_time = time.time()
        outputs = model(images)
        end_time = time.time()
        inference_times.append((end_time - start_time) / images.size(0))
        _, preds = outputs.max(1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Calculations
avg_inf_time = np.mean(inference_times) * 1000 # ms
model_size = os.path.getsize(best_model_path) / (1024 * 1024)

# Summary Table
metrics_dict = {
    "Metric": ["Test Accuracy", "Precision (Macro)", "Recall (Macro)", "F1-Score", "Inf. Time/Img", "Model Size"],
    "Value": [
        f"{100 * (np.array(all_preds) == np.array(all_labels)).mean():.2f}%",
        f"{100 * precision_score(all_labels, all_preds, average='macro'):.2f}%",
        f"{100 * recall_score(all_labels, all_preds, average='macro'):.2f}%",
        f"{100 * f1_score(all_labels, all_preds, average='macro'):.2f}%",
        f"{avg_inf_time:.3f} ms",
        f"{model_size:.2f} MB"
    ]
}
display(pd.DataFrame(metrics_dict))

# Confusion Matrix

plt.figure(figsize=(8, 6))
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=train_dataset.classes, yticklabels=train_dataset.classes)
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nDetailed Classification Report:\n")
print(classification_report(all_labels, all_preds, target_names=train_dataset.classes))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from tqdm.auto import tqdm
import os
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report, \
    f1_score, precision_score, recall_score

# -----------------------------
# 1. Device & Paths
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

base_path = "/content/dataset/classification"
train_dir = os.path.join(base_path, "train")
val_dir   = os.path.join(base_path, "val")
test_dir  = os.path.join(base_path, "test")

models_dir = "/content/drive/MyDrive/SmartVision AI - Intelligent Multi-Class Object Recognition System/models"
os.makedirs(models_dir, exist_ok=True)
best_model_path = os.path.join(models_dir, "best_efficientnet_b3.pth")

# -----------------------------
# 2. Transforms (EfficientNet-B3)
# -----------------------------
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(300, scale=(0.85, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

val_test_transform = transforms.Compose([
    transforms.Resize(320),
    transforms.CenterCrop(300),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# -----------------------------
# 3. Datasets & Loaders
# (num_workers=0 fixes your crash)
# -----------------------------
train_dataset = ImageFolder(train_dir, transform=train_transform)
val_dataset   = ImageFolder(val_dir, transform=val_test_transform)
test_dataset  = ImageFolder(test_dir, transform=val_test_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True,
                          num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False,
                          num_workers=0, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False,
                          num_workers=0, pin_memory=True)

num_classes = len(train_dataset.classes)

# -----------------------------
# 4. Model: EfficientNet-B3
# -----------------------------
model = models.efficientnet_b3(weights="IMAGENET1K_V1")

# Replace classifier
model.classifier[1] = nn.Linear(
    model.classifier[1].in_features, num_classes
)

model = model.to(device)

# -----------------------------
# 5. Phase 1: Train Classifier Only
# -----------------------------
for param in model.parameters():
    param.requires_grad = False

for param in model.classifier.parameters():
    param.requires_grad = True

criterion = nn.CrossEntropyLoss()

optimizer = optim.AdamW(
    model.classifier.parameters(), lr=1e-3, weight_decay=1e-4
)

print("\nðŸš€ Phase 1: Training classifier only")

for epoch in range(3):
    model.train()
    correct, total = 0, 0

    for images, labels in tqdm(train_loader, desc=f"Phase1 Epoch {epoch+1}/3"):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, preds = outputs.max(1)
        total += labels.size(0)
        correct += preds.eq(labels).sum().item()

    print(f"Train Acc: {100 * correct / total:.2f}%")

# -----------------------------
# 6. Phase 2: Full Fine-Tuning
# -----------------------------
for param in model.parameters():
    param.requires_grad = True

optimizer = optim.AdamW([
    {"params": model.features.parameters(), "lr": 3e-5},
    {"params": model.classifier.parameters(), "lr": 3e-4}
], weight_decay=1e-4)

num_epochs = 5
scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=num_epochs, eta_min=1e-6
)

best_val_acc = 0.0

print("\nðŸ”¥ Phase 2: Full fine-tuning")

for epoch in range(num_epochs):
    # ---- Train ----
    model.train()
    train_correct, train_total = 0, 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, preds = outputs.max(1)
        train_total += labels.size(0)
        train_correct += preds.eq(labels).sum().item()

    train_acc = 100 * train_correct / train_total

    # ---- Validation ----
    model.eval()
    val_correct, val_total = 0, 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = outputs.max(1)
            val_total += labels.size(0)
            val_correct += preds.eq(labels).sum().item()

    val_acc = 100 * val_correct / val_total

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), best_model_path)
        print(f"ðŸŒŸ New Best Model Saved: {val_acc:.2f}%")

    print(f"Epoch [{epoch+1}/{num_epochs}] | "
          f"Train: {train_acc:.2f}% | Val: {val_acc:.2f}%")

    scheduler.step()

# -----------------------------
# 7. Evaluation on Test Set
# -----------------------------
print("\nðŸ“Š FINAL EVALUATION")

model.load_state_dict(torch.load(best_model_path))
model.eval()

all_preds, all_labels, inference_times = [], [], []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)

        start = time.time()
        outputs = model(images)
        end = time.time()

        inference_times.append((end - start) / images.size(0))
        _, preds = outputs.max(1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Metrics
test_acc = 100 * (np.array(all_preds) == np.array(all_labels)).mean()
precision = 100 * precision_score(all_labels, all_preds, average="macro")
recall = 100 * recall_score(all_labels, all_preds, average="macro")
f1 = 100 * f1_score(all_labels, all_preds, average="macro")
avg_inf_time = np.mean(inference_times) * 1000
model_size = os.path.getsize(best_model_path) / (1024 * 1024)

df = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1", "Inference (ms)", "Model Size (MB)"],
    "Value": [f"{test_acc:.2f}%", f"{precision:.2f}%", f"{recall:.2f}%",
              f"{f1:.2f}%", f"{avg_inf_time:.2f}", f"{model_size:.2f}"]
})

display(df)

# -----------------------------
# 8. Confusion Matrix
# -----------------------------
plt.figure(figsize=(10, 8))
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=train_dataset.classes,
            yticklabels=train_dataset.classes)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("\nClassification Report:\n")
print(classification_report(all_labels, all_preds,
                            target_names=train_dataset.classes))
```
